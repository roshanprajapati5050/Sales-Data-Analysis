## Creating Charts for Data Visualization

### Plothing Profit Over Time

```python
df['Date'] = pd.to_datetime(df['Date'])  # Ensure Date is in datetime format
df.groupby('Date')['Profit'].sum().plot(figsize=(10,5), title='Profit Over Time')
plt.ylabel('Total Profit')
plt.show()
```

### Yearly Revenue Trends

```python
Yearly = df.groupby('Year')['Revenue'].sum().sort_index()
sns.barplot(x=Yearly.index, y=Yearly.values)
plt.title('Yearly Revenue')
plt.xlabel('Years')
plt.ylabel('Revenue')
plt.show()
```

### Age Group Distribution using countplot

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Age_Group')
plt.title('Customer Age Group Distribution')
plt.show()
```

### Gender-wise Revenue

```python
gender_revenue = df.groupby('Gender')['Revenue'].sum()

plt.pie(gender_revenue, labels=gender_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue by Gender')
plt.axis('equal')  # Makes the pie chart a circle
plt.show()
```

### **Pie Chart:** Breakdown of Cost, Profit (highlighted), and Revenue proportions.

```python
# Total values
totals = df[['Cost', 'Profit', 'Revenue']].sum()

# Labels, colors, and explode settings
labels = ['Cost', 'Profit', 'Revenue']
colors = ['red', 'green', 'orange']
explode = [0, 0.05, 0]  # Slightly pull out the Profit slice

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(totals, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=90, explode=explode, shadow=True)
plt.title('Overall Financial Distribution (Profit Highlighted)')
plt.axis('equal')
plt.show()
```

### Top 10 States by Profit

```python
top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
top_states.plot(kind='bar', title='Top 10 States by Profit')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```

### Country-Wise Revenue Pie Chart

```python
df.groupby('Country')['Revenue'].sum().plot.pie(autopct='%1.1f%%')
plt.title('Revenue by Country')
plt.ylabel('')
plt.show()
```

### Histogram: Shows how customer ages are distributed across the dataset. 
### Bar Chart: Compares total revenue generated by each gender.
### Line Chart: Displays profit trends over time.
### Box Plot: Highlights profit variation across product categories.

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 rows, 2 columns

# Histogram of Customer Age (Distribution)
axs[0, 0].hist(df['Customer_Age'], bins=10, color='orange', edgecolor='black')
axs[0, 0].set_title('Customer Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')

# Bar Chart: Revenue by Gender
revenue_by_gender = df.groupby('Gender')['Revenue'].sum()
axs[0, 1].bar(revenue_by_gender.index, revenue_by_gender.values, color='teal')
axs[0, 1].set_title('Revenue by Gender')
axs[0, 1].set_xlabel('Gender')
axs[0, 1].set_ylabel('Revenue')

# Line Chart: Profit Over Time (if Date exists)
df_sorted = df.sort_values('Date')  # Sort by date to show trend
axs[1, 0].plot(df_sorted['Date'], df_sorted['Profit'], color='crimson', linewidth=1.5)
axs[1, 0].set_title('Profit Trend Over Time')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Profit')

# Box Plot: Profit Distribution by Product Category
df.boxplot(column='Profit', by='Product_Category', ax=axs[1, 1])
axs[1, 1].set_title('Profit by Product Category')
axs[1, 1].set_xlabel('Product Category')
axs[1, 1].set_ylabel('Profit')
axs[1, 1].tick_params(axis='x', rotation=45)
fig.suptitle('')  # Optional: Remove the default boxplot title

plt.tight_layout()
plt.show()
```
